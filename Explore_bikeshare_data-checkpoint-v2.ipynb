### Question 1

> What is the most common start station in each city?

> Is this start station the same every day of the week?

```r
#Checking the most commonly used start station for Chicago

start_stations_c <- chi %>%   
  group_by(Start.Station) %>%
  summarise(n = n()) %>%
  filter(n > 40)


 ggplot(start_stations_c, aes(Start.Station, n))+
  geom_col()+
  labs(title = 'Chicago')+
    coord_flip() 
```

```r
# Sort descending order and show the top five
start_stations_c <- start_stations_c %>% arrange(desc(n))
head(start_stations_c, 5)
```

```r
#Checking the most commonly used start station for New York
start_stations_n <- ny %>%   
  group_by(Start.Station) %>%
  summarise(n = n()) %>%
  filter(n > 200)

ggplot(start_stations_n, aes(Start.Station, n))+
  geom_col()+
  labs(title = 'New York')+
    coord_flip()
```

```r
# Sort descending order and show the top five
start_stations_n <- start_stations_n %>% arrange(desc(n))
head(start_stations_n, 5)
```

```r
#Checking the most commonly used start station for Washington

start_stations_w <- wash %>%   
  group_by(Start.Station) %>%
  summarise(n = n()) %>%
  filter(n > 500)

ggplot(start_stations_w, aes(Start.Station, n))+
  geom_col()+
  labs(title = 'Washington')+
  coord_flip()
     
```

```r
# Sort descending order
start_stations_w <- start_stations_w %>% arrange(desc(n))
head(start_stations_w, 5)
```

```r
#Reducing the dataframes to work with them better
chi_ss <- chi[c("X", "City", "Start.Station")]
wash_ss <- wash[c("X", "City", "Start.Station")]
ny_ss <- ny[c("X", "City", "Start.Station")]

# Merge the selected columns into a single dataframe
ss_merged_df <- rbind(chi_ss, wash_ss, ny_ss)
head(ss_merged_df, 10)
```

```r


# Calculate the counts of start stations by city
start_stations_counts <- ss_merged_df %>%
  group_by(City, Start.Station) %>%
  summarise(Count = n()) %>%
  filter(!is.na(Start.Station) & Start.Station != "") %>%
  arrange(City, desc(Count))


# Select the top three start stations for each city
top_start_stations <- start_stations_counts %>%
  group_by(City) %>%
  top_n(n = 3, wt = Count)

# using ggplot to chart the top three start stations in each city
ggplot(top_start_stations, aes(x = City, y = Count, fill = Start.Station)) +
  geom_bar(stat = "identity", position = "dodge") +
  geom_text(aes(label = Count), position = position_dodge(width = 0.9), vjust = -0.5, color = "black", size = 3) +  
  labs(title = "Top Three Start Stations by City",
       x = "City",
       y = "Count") +
  theme_minimal()
```

###### Results
The most often used starting station in each city is:

>Chicago - Streeter Dr. & Grand Ave

>New York - Pershing Square North

>Washington - Columbus Circle/Union Station

```r

# In order to find out if the most popular start station remains popular throughout the week, we need get the day of the week
chi <- chi %>%
  mutate(Start.Day = format(as.Date(Start.Time), "%A"))

wash <- wash %>%
  mutate(Start.Day = format(as.Date(Start.Time), "%A"))

ny <- ny %>%
  mutate(Start.Day = format(as.Date(Start.Time), "%A"))

# Find the most often used Start Station for each day of the week in each dataframe
chi_most_often <- chi %>%
  group_by(Start.Day, Start.Station) %>%
  tally() %>%
  group_by(Start.Day) %>%
  top_n(1, wt = n)

wash_most_often <- wash %>%
  group_by(Start.Day, Start.Station) %>%
  tally() %>%
  group_by(Start.Day) %>%
  filter(!is.na(Start.Station) & Start.Station != "") %>%
  top_n(1, wt = n)


ny_most_often <- ny %>%
  group_by(Start.Day, Start.Station) %>%
  tally() %>%
  group_by(Start.Day) %>%
  top_n(1, wt = n)

# Print the results
chi_most_often
wash_most_often
ny_most_often
```

```r
# Create the bar chart to display most often used start station by day of the week
bar_chart_chi <- ggplot(chi_most_often, aes(x = Start.Day, y = n, fill = Start.Station)) +
  geom_bar(stat = "identity", position = "dodge") +
  geom_text(aes(label = n), vjust = -0.5, color = "black", size = 3) +  
  labs(title = "Most Often Used Start Stations in Chicago by Day",
       x = "Day of the Week",
       y = "Count") +
  theme_minimal()

print(bar_chart_chi)
```

```r
# Create the bar chart to display most often used start station by day of the week
bar_chart_wash <- ggplot(wash_most_often, aes(x = Start.Day, y = n, fill = Start.Station)) +
  geom_bar(stat = "identity", position = "dodge") +
  geom_text(aes(label = n), vjust = -0.5, color = "black", size = 3) + 
  labs(title = "Most Often Used Start Stations in Washington by Day",
       x = "Day of the Week",
       y = "Count") +
  theme_minimal()

print(bar_chart_wash)
```

```r
# Create the bar chart to display most often used start station by day of the week
bar_chart_ny <- ggplot(ny_most_often, aes(x = Start.Day, y = n, fill = Start.Station)) +
  geom_bar(stat = "identity", position = "dodge") +
  geom_text(aes(label = n), vjust = -0.5, color = "black", size = 3) + 
  labs(title = "Most Often Used Start Stations in New York by Day",
       x = "Day of the Week",
       y = "Count") +
  theme_minimal()

print(bar_chart_ny)
```

### Question 2

>What is the most common age of riders in each city (excluding Washington)? 

>Does this vary when considering gender?

```r
#calculating age from birth year and creating a range for the ages
calculate_age_range <- function(data) {
  data %>%
    mutate(Age = floor((as.integer(format(Sys.Date(), "%Y")) - Birth.Year) / 10) * 10) %>%
    group_by(Age) %>%
    summarise(count = n()) %>%
    filter(Age >= 10 & Age < 100) %>% #filtered out outlier ages
    arrange(desc(count))
}

# Calculate the most common age range counting by 10 for each dataframe
ny_age_range <- calculate_age_range(ny)
chi_age_range <- calculate_age_range(chi)

ny_age_range
chi_age_range
```

```r
create_bar_chart <- function(data, city) {
  ggplot(data, aes(x = Age, y = count, fill = factor(Age))) +
    geom_bar(stat = "identity") +
    geom_text(aes(label = count), vjust = -0.5, color = "black", size = 3) +
    labs(title = paste("Most Common Age Range in", city),
         x = "Age Range",
         y = "Count") +
    theme_minimal()
}

# Create a bar chart with data labels for ny_age_range
bar_chart_ny <- create_bar_chart(ny_age_range, "New York")

# Create a bar chart with data labels for chi_age_range
bar_chart_chi <- create_bar_chart(chi_age_range, "Chicago")

# Print the bar charts
print(bar_chart_ny)
print(bar_chart_chi)
```

```r
#referenced: https://dplyr.tidyverse.org/reference/mutate.html
#calculating age from birth year and creating a range for the ages
calculate_age_range <- function(data) {
  data %>%
    mutate(Age = floor((as.integer(format(Sys.Date(), "%Y")) - Birth.Year) / 10) * 10) %>%
    group_by(Age, Gender) %>%
    summarise(count = n()) %>%
    filter(Age >= 10 & Age < 100) %>% #filtered out outlier ages
    arrange(desc(count))
}

# Calculate the most common age range counting by 10 for each dataframe
ny_age_range <- calculate_age_range(ny)
chi_age_range <- calculate_age_range(chi)

ny_age_range
chi_age_range
```

```r
#displaying age ranges by gender 
ggplot(ny_age_range, aes(x = factor(Age), y = count, fill = Gender)) +
  geom_bar(stat = "identity", position = "dodge") +
  labs(title = "Age Range by Gender - NY",
       x = "Age Range",
       y = "Count",
       fill = "Gender") +
  theme_minimal()
```

```r
# Bar chart for chi_age_range
ggplot(chi_age_range, aes(x = factor(Age), y = count, fill = Gender)) +
  geom_bar(stat = "identity", position = "dodge") +
  labs(title = "Age Range by Gender - CHI",
       x = "Age Range",
       y = "Count",
       fill = "Gender") +
  theme_minimal()
```

### Question 3

> What time of day is most common in each city?

> What is the most common end time in each city?

```r
#create new column called start.hour

create_start_hour_column <- function(data) {
  data %>%
    mutate(Start.Time = ymd_hms(Start.Time),  # Convert Start.Time to proper date/time format
           start.hour = hour(Start.Time))     # Extract hour component
}

wash <- create_start_hour_column(wash)
ny <- create_start_hour_column(ny)
chi <- create_start_hour_column(chi)

wash
```

```r
#finding the most popular start hour
find_most_popular_start_hour <- function(data) {
  data %>%
    group_by(X, City, start.hour) %>%
    summarise(count = n()) %>%
    filter(count == max(count))
}

wash_most_popular_start_hour <- find_most_popular_start_hour(wash)
ny_most_popular_start_hour <- find_most_popular_start_hour(ny)
chi_most_popular_start_hour <- find_most_popular_start_hour(chi)

wash_most_popular_start_hour 
ny_most_popular_start_hour  
chi_most_popular_start_hour
```

```r
#creating a merged dataframe to work with - plotting the above data was cumbersome
start_hr_merge <- rbind(chi, ny)

#need to drop the gender and birth year columns to merge with wash
start_hr_merge <- start_hr_merge %>%
  select(-c("Gender", "Birth.Year"))

#complete the merge
start_hr_merge <- rbind(start_hr_merge, wash)

start_hr_merge
```

```r
#find the top start time for each city in the merged dataframe


top_start_times <- start_hr_merge %>%
  group_by(City, start.hour) %>%
  count() %>%
  na.omit() %>%
  top_n(1, wt = n) %>%
  arrange(City, desc(n))

# Find the most common start hour for each city
most_common_start_hour <- top_start_times %>%
  group_by(City) %>%
  filter(n == max(n)) %>%
  distinct(City, .keep_all = TRUE)

most_common_hour
```

```r
# Plot the top start times for each city on a single bar plot
ggplot(top_start_times, aes(x = start.hour, y = n, color = City, group = City)) +
  geom_line() +
  geom_point() +
  geom_text(data = most_common_start_hour, aes(label = start.hour),
            vjust = -0.5, color = "black", size = 3) + 
  labs(title = "Top Start Times for Each City",
       x = "Start Hour",
       y = "Count",
       color = "City") +
  theme_minimal()
```

```r
#create new column called end.hour

create_end_hour_column <- function(data) {
  data %>%
    mutate(End.Time = ymd_hms(End.Time),  
           stop.hour = hour(End.Time))     
}

wash <- create_end_hour_column(wash)
ny <- create_end_hour_column(ny)
chi <- create_end_hour_column(chi)

wash
```

```r
#creating a new merged dataframe to work with
stop_hr_merge <- rbind(chi, ny)
stop_hr_merge


#need to drop the gender and birth year columns to merge with wash
stop_hr_merge <- stop_hr_merge %>%
  select(-c("Gender", "Birth.Year"))


#complete the merge
stop_hr_merge <- rbind(stop_hr_merge, wash)

stop_hr_merge
```

```r
#find the top start time for each city in the merged dataframe

top_stop_times <- stop_hr_merge %>%
  group_by(City, stop.hour) %>%
  count() %>%
  na.omit() %>%
  top_n(1, wt = n) %>%
  arrange(City, desc(n))

# Find the most common start hour for each city
most_common_stop_hour <- top_stop_times %>%
  group_by(City) %>%
  filter(n == max(n)) %>%
  distinct(City, .keep_all = TRUE)

most_common_stop_hour
```

```r
# Plot the top start times for each city on a single bar plot
ggplot(top_stop_times, aes(x = stop.hour, y = n, color = City, group = City)) +
  geom_line() +
  geom_point() +
  geom_text(data = most_common_stop_hour, aes(label = stop.hour),
            vjust = -0.5, color = "black", size = 3) + 
  labs(title = "Top Stop Times for Each City",
       x = "Stop Hour",
       y = "Count",
       color = "City") +
  theme_minimal()
```
